```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import chromedriver_autoinstaller
```

```python
chromedriver_autoinstaller.install()
browser = webdriver.Chrome()
```

```python
browser.get('https://www.trip.com/hotels/list?city=19961&checkin=2023-06-12&checkout=2023-06-13')
```

```python
soup=BeautifulSoup(browser.page_source,'lxml')
```

```python
rows=soup.find_all('div',class_='hotel-info')
```

```python
data = []

for row in rows:
    nama_element = row.find(class_='name')
    nama = nama_element.text if nama_element is not None and nama_element.text.strip() != "" else None

    location_element = row.find(class_='list-card-transport-v8')
    location = location_element.text if location_element is not None and location_element.text.strip() != "" else None

    penawaran_terbaik_element = row.find(class_='room-panel-roominfo')
    penawaran_terbaik = penawaran_terbaik_element.text if penawaran_terbaik_element is not None and penawaran_terbaik_element.text.strip() != "" else None

    rating_element = row.find(class_='score')
    rating = rating_element.text if rating_element is not None and rating_element.text.strip() != "" else None

    tax_element = row.find(class_='not-break')
    tax = tax_element.text if tax_element is not None and tax_element.text.strip() != "" else None

    pengunjung_web_element = row.find(class_='count')
    pengunjung_web = pengunjung_web_element.text if pengunjung_web_element is not None and pengunjung_web_element.text.strip() != "" else None
    
    dt = {
        'nama': nama,
        'location': location,
        'penawaran_terbaik': penawaran_terbaik,
        'rating': rating,
        'tax': tax,
        'pengunjung_web': pengunjung_web
    }

    data.append(dt)

```

```python
df=pd.DataFrame(data)
```
